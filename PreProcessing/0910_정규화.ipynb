```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('0910 착과수_4.csv')
```

```python
df
```

```python
df_copy = df.copy()

df_copy['STRG_DT'] = pd.to_datetime(df_copy['STRG_DT'])

# 'Date' 열을 인덱스로 설정
df_copy.set_index('STRG_DT', inplace=True)
```

```python
#standard scaler
scaler = StandardScaler()

# 데이터프레임의 모든 열에 Z-Score 정규화 적용
normalized_df = scaler.fit_transform(df_copy)

# 정규화된 데이터를 다시 데이터프레임으로 변환
normalized_df = pd.DataFrame(normalized_df, columns=df_copy.columns)

```

```python
normalized_df['STRG_DT'] = df['STRG_DT']
```

```python
normalized_df
```

```python
# 'Date' 열을 'normalized_df'의 첫 번째 열로 이동
normalized_df = normalized_df[['STRG_DT'] + [col for col in normalized_df if col != 'STRG_DT']]

normalized_df
```

```python
normalized_df.to_csv('0910 착과수_normalized.csv', index = False, header = True)
```

```python
#robust scaler 사용 (이상치에 강함)
scaler = RobustScaler()

# 데이터프레임의 모든 열을 RobustScaler로 정규화
df_robust = scaler.fit_transform(df)

# 정규화된 데이터를 다시 데이터프레임으로 변환
df_robust = pd.DataFrame(df_robust, columns=df.columns)

df_robust
```

```python
scaler = MinMaxScaler()

# 데이터프레임의 모든 열을 Min-Max 스케일러로 정규화
df_minmax = scaler.fit_transform(df)

# 정규화된 데이터를 다시 데이터프레임으로 변환
df_minmax = pd.DataFrame(df_minmax, columns=df.columns)

df_minmax
```
