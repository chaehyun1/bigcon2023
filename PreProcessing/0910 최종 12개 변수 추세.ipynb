```python
FRUIT_SETTING['STRG_DT'] = pd.to_datetime(FRUIT_SETTING['STRG_DT'])

# 그래프 그리기
for i in range(1, len(fruit_setting)):
    FRUIT_SETTING.plot(x='STRG_DT', y=fruit_setting[i], kind='line', figsize=(15, 10))
    plt.title(fruit_setting[i])
    plt.grid(True)
    plt.show()
```
