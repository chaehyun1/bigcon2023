```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import os
import seaborn as sns
```

```python
FRUIT_SETTING = pd.read_csv('C:/Users/chaeh/bigcon/dataset/생육변수와착과수.csv')
```

```python
FRUIT_SETTING.head()
```

```python
colormap = plt.cm.PuBu
plt.figure(figsize=(20, 15))
sns.heatmap(FRUIT_SETTING.corr(), linewidths = 0.1, vmax = 1.0, square = True, 
            cmap = colormap, linecolor = 'white', annot = True, annot_kws = {'size':12})
plt.show()
```

```python
FRUIT_SETTING['STRG_DT'] = pd.to_datetime(FRUIT_SETTING['STRG_DT'])
FRUIT_SETTING['STRG_DT'] = FRUIT_SETTING['STRG_DT'].dt.date
```

```python
FRUIT_SETTING.head()
```

```python
FRUIT_SETTING.set_index('STRG_DT', inplace=True)
```

```python
daily_grouped = FRUIT_SETTING.groupby(FRUIT_SETTING.index).mean()
daily_grouped
```

```python
daily_grouped.reset_index(inplace=True)
daily_grouped
```

```python
# 날짜 컬럼을 월 정보로 변환
daily_grouped['STRG_DT'] = pd.to_datetime(daily_grouped['STRG_DT'])
daily_grouped['month_only'] = daily_grouped['STRG_DT'].dt.month
daily_grouped
```

```python
daily_grouped['month_only'] = daily_grouped['month_only'].astype(int)
daily_grouped
```
