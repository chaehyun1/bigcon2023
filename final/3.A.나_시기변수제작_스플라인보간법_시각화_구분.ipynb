```python
import pandas as pd
df= pd.read_csv("C:/Users/user/Desktop/빅콘테스트/GROUPED_GROWTH_DATA.csv")
print(df.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import os
import seaborn as sns
```

```python
# cubic spline interpolation#삼차다항식 #데이터 네개 point필요할 것.
 
#더 차수 높은 거 쓰면 더 예측오차 작아질 것. 아마  #그래도 오차 발생위험은 항상 있음.
from scipy.interpolate import CubicSpline

x_fine = np.linspace(0, 181, num=100)

spl = CubicSpline(df.index, df["FRST_TREE_CNT"])
y_cubic_spline = spl(x_fine)
spl = CubicSpline(df.index, df["LEAF_LNGTH"])
y_cubic_spline_1 = spl(x_fine)
spl = CubicSpline(df.index, df["LEAF_CNT"])
y_cubic_spline_2 = spl(x_fine)

plt.figure(figsize=(9,6))
#plt.plot(x_fine, y_cubic_spline, 'b-', label='FRST_TREE_CNT')
plt.plot(x_fine, y_cubic_spline_1, 'r-', label="LEAF_LNGTH")
plt.plot(x_fine, y_cubic_spline_2, 'y-', label="LEAF_CNT")
plt.axvline(64, color= "green",ls="--")
plt.axvline(137, color= "green",ls="--")
#plt.plot(df.index, df["FRST_TREE_CNT"], 'b', ms=1, label='FRST_TREE_CNT')
#plt.plot(df.index, df["LEAF_LNGTH"], 'r-',ms=1, label='LEAF_LNGTH')
#plt.plot(df.index, df["LEAF_CNT"], 'y-', ms=1, label='LEAF_CNT')
#plt.plot(x_fine, y_true, 'g--', label=f'$\cos(-x^2/{n})$')
plt.legend(loc='best')
plt.show()
```
